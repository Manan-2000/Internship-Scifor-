```python
1- Multi-Layer Perceptron is also called known as Artificial Neural Network is a neural network, it has an input layer an output layer and
a hidden layer. Single-layer perceptron has only have input layer and an output layer there are multiple neurons in all the layers of
Multi-Layer Perceptron. But in single-layer perceptron we have only one neuron so its very simple and can not handle complex problems and
it was not able to learn, but MLP can.
```

```python
2- In MLP there are inputs they are send to each neuron of the input layer with some weights to them,x-net is calculated and then they have
activation funtions in them, the final output of all neurons of input layer goes to all the neurons of hidden layer with some weights added
to the outputs,similar x-net is calculated and activation function is applied. Then the output of the hidden layer is added to the output
layer with some weights. Same the output is added to the output layer with some weights.x-net is calculated and an activation funtion is
applied, then the final output is recieved.
```

```python
3- Weights are asssigned randomly using distributions like normal or uniform distributions. there are many ways to initialize weights.
Xavier initialization,He initialization,Normal initialization. They are important because the weights are assinged randomly if there
were no initializers then the weights may be too low which will cause vanishing gradient problem or if the weights are assigned too high
we have exploding gradient problem or that no 2 weights are equal.
```

```python
4- Main purpose of Activation Functions is to make the x-net that is we get from the sum of inputs and weights and bias is passed as an
output to the next layer. Some commonly used activation functions are:
Sigmoid(used for binary classsification),
Tan h(used for regression),
ReLU(used mostly in hidden layer),
Leaky ReLU, (used to address dying ReLU problem)
Parameterized ReLU,
Exponential ReLU,
Swish, (combines the positives of Sigmoid and ReLU)
Softmax(used for multiclass classification).
```

```python
5- In backpropagation we use the loss that we get from the loss function and go back from the output layer to the hidden layer to the
input layer while updating the weights using optimizers such as gradien descent, adagrad, adadelta, adam. As we keep updating the weights
the loss keeps decreasing which makes the performance of the model better and loss is minimimal.
```

```python
6- There is no exact way to find how many neurons or hidden layers there should be, you can have as many as you want but it is prefered
to be as minimal as possible while getting the best performance. Increasing the layers or the neurons increases the computations meaning
increase in time and may casue overfitting.
```

# Load MNIST Dataset and build a Deep Learning Model.

### Things to be done:
    1- Load the MNIST Data
    2- Divide the data into train and test sets
    3- Plot the first 10 images using imshow()
    4- Convert images to categorical format
    5- Create a sequential model
    6- Plot the model summary
    7- Plot the architecture of the model with its layer shapes
    8- Compile the model
    9- Fit the model with 20 epochs (as per your desktop configuration)
    10- Plot Training and Testing Loss
    11- Save the model in .h5 extension
    12- Load the model
    13- Test the model

```python
import numpy as np
import matplotlib.pyplot as mpl
from keras.layers import Dense,Flatten
from keras.models import Sequential
from keras.utils import to_categorical
from keras import utils
from keras.datasets import mnist
import tensorflow as tf
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data() #load MNIST dataset
```

```python
fig, axes = mpl.subplots(ncols=10,sharex=False,sharey=True,figsize=(10,4))
for i in range(10):
  axes[i].set_title(y_train[1])
  axes[i].imshow(x_train[i],cmap='gray')
  axes[i].get_xaxis().set_visible(False)
  axes[i].get_yaxis().set_visible(False)
mpl.show()  #plot the first 10 images using imshow()
```

```python
#convert images to categorical format
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
mod=Sequential() #creating a sequential model
mod.add(Flatten(input_shape=(28,28)))
mod.add(Dense(10,activation='sigmoid'))
mod.add(Dense(5,activation='ReLU'))
mod.add(Dense(10,activation='softmax'))
mod.summary() #plotting model summary
```

```python
utils.plot_model(mod,show_shapes=True) #plot the architecture of the model with its layer shapes
```

```python
mod.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['acc']) #compiling the model
```

```python
res=mod.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test)) #fitting the model
```

```python
mpl.plot(res.history['loss'],label='Train Loss')
mpl.plot(res.history['val_loss'],label='Validation Loss')
mpl.legend() #plot for training and testing loss
```

```python
mod.save('MNIST_Manan.h5') #saving the model
```

```python
model=tf.keras.models.load_model('MNIST_Manan.h5') #loading the model
```

```python
mpl.imshow(x_test[11],cmap='gray')
```

```python
x=np.reshape(x_test[11],(1,28,28))
predicted_label = np.argmax(model.predict(x)) #testing the model
print(predicted_label)
```
