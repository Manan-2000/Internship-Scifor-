```python
import numpy as np
import pandas as pd
import re
```

```python
df=pd.DataFrame(pd.read_csv('/content/movies.csv'))
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['Gross','ONE-LINE'], axis=1, inplace=True)
print(df.shape)
```

```python
df_dup = df[df.duplicated(keep=False)]
print(df_dup.shape)
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.drop_duplicates(subset ="MOVIES",keep = False, inplace = True)
df.shape
```

```python
df.isna().sum()
```

```python
df.rename(columns={'RunTime': 'RUNTIME'}, inplace=True)
```

```python
filt= (df['YEAR'].isna()) & \
    (df['RATING'].isna()) & \
    (df['VOTES'].isna()) & \
    (df['RUNTIME'].isna())

df_na = df[filt]
df_na.shape
```

```python
df_na.head()
```

```python
df['GENRE'] = df['GENRE'].str.replace('\n', '')
df['STARS'] = df['STARS'].str.strip()
df['STARS'] = df['STARS'].str.replace('\n', '')
df['YEAR'] = df['YEAR'].str.replace('\n', '')
df['VOTES'] = df['VOTES'].str.replace('\n', '')
#df["STARS"].head(10)
df
```

```python
df['VOTES'] = df['VOTES'].str.replace(',', '').astype(float)
```

```python
mean_votes = df['VOTES'].mean()
df['VOTES'].fillna(mean_votes, inplace=True)
df.isnull().sum()
```
