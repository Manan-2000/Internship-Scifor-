```python
import seaborn as sb
d=sb.load_dataset("car_crashes")

#How many car crashes was because of alcohol
sb.violinplot(x="total",y="alcohol",data=d)
```

```python
#How many crashes was caused by speeding
sb.boxplot(x="total",y="speeding",data=d)
```

```python
#Comparision of car crashes between speeding and alcohol
sb.scatterplot(x='total',y='alcohol',data=d)
sb.scatterplot(x='total',y='speeding',data=d)
```
