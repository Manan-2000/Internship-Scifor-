```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import accuracy_score , confusion_matrix ,roc_auc_score,roc_curve,classification_report

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/diabetes.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
plt.figure(figsize=(20,15))
graph=1
for column in data:
    if graph <=9:
        ax=plt.subplot(3,3,graph)
        sb.distplot(data[column])
        plt.xlabel(column)
    graph +=1
plt.tight_layout()
```

```python
data['BMI']=data['BMI'].replace(0,data['BMI'].mean())
data['Insulin']=data['Insulin'].replace(0,data['Insulin'].mean())
data['BloodPressure']=data['BloodPressure'].replace(0,data['BloodPressure'].mean())
data['Glucose']=data['Glucose'].replace(0,data['Glucose'].mean())
data['SkinThickness']=data['SkinThickness'].replace(0,data['SkinThickness'].mean())
```

```python
plt.figure(figsize=(20,15))
plotnumber=1
for column in data:
    if plotnumber <=9:
        ax=plt.subplot(3,3,plotnumber)
        sb.distplot(data[column])
        plt.xlabel(column,fontsize=20)
    plotnumber +=1
plt.tight_layout()
```

```python
df_feature=data.drop('Outcome',axis=1)
plt.figure(figsize=(20,15),facecolor='green')
plotnumber=1
for column in df_feature:
    if plotnumber <=9:
        plt.subplot(3,3,plotnumber)
        ax=sb.boxplot(data=df_feature[column])
        plt.xlabel(column,fontsize=15)
    plotnumber +=1
plt.tight_layout()
```

```python
q1=data.quantile(0.25)
q3=data.quantile(0.75)
iqr =q3  - q1
```

```python
pred_high = (q3.Pregnancies + (1.5 * iqr.Pregnancies))
pred_high
```

```python
index=np.where(data['Pregnancies']>pred_high)
index
```

```python
data=data.drop(data.index[index])
data.reset_index()
```

```python
bp_high =(q3.BloodPressure + (1.5 * iqr.BloodPressure))
print(bp_high)
index=np.where(data['BloodPressure']>bp_high)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
st_high=(q3.SkinThickness + (1.5 * iqr.SkinThickness))
print(st_high)
index=np.where(data['SkinThickness']>st_high)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
insu_high=(q3.Insulin + (1.5 * iqr.Insulin))
print(insu_high)
index=np.where(data['Insulin']>insu_high)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
bmi_high=(q3.BMI + (1.5 * iqr.BMI))
print(bmi_high)
index=np.where(data['BMI']>bmi_high)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
dpf_high=(q3.DiabetesPedigreeFunction + (1.5 *iqr.DiabetesPedigreeFunction))
print(dpf_high)
index=np.where(data['DiabetesPedigreeFunction']>dpf_high)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
age_high = (q3.Age + (1.5*iqr.Age))
print(age_high)
index=np.where(data['Age']>age_high)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
bp_low=(q1.BloodPressure - (1.5 * iqr.BloodPressure))
print(bp_low)
index=np.where(data['BloodPressure']<bp_low)
data=data.drop(data.index[index])
print(data.shape)
data.reset_index()
```

```python
plt.figure(figsize=(20,15))
plotnumber = 1
for column in data :
    if plotnumber <=9:
        ax=plt.subplot(3,3,plotnumber)
        sb.distplot(data[column])
        plt.xlabel(column,fontsize=20)
    plotnumber +=1
plt.tight_layout()
```

```python
x=data.drop(columns=['Outcome'])

y=data.Outcome
```

```python
plt.figure(figsize=(15,20))
plotnumber = 1
for column in x:
    if plotnumber <=9:
        ax=plt.subplot(3,3,plotnumber)
        sb.stripplot(x=y,y=x[column])
    plotnumber +=1
plt.show()
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled.shape
```

```python
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i) for i in range (x_scaled.shape[1])]
vif['Features']=x.columns
vif
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=355)
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
y_pred=log_reg.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
conf_mat=confusion_matrix(y_test,y_pred)
conf_mat
```

```python
fpr,tpr,threshold =roc_curve(y_test,y_pred)
print('Threshold = ',threshold)
print('True Positive Rate = ',tpr)
print('False Positive Rate = ',fpr)
```

```python
plt.plot(fpr,tpr,color='orange',label='Roc')
plt.plot([0,1],[0,1],color='darkblue',linestyle='--')
plt.xlabel('Fasle Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('RecieVer Operating Characterstics (ROC)')
plt.legend()
plt.show()
```
