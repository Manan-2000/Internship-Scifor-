```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('/content/class.csv')
df
```

```python
plt.scatter(df.students,df.marks)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['students']],df.marks)
```

```python
reg.predict([[36]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
0.74678112*36+68.70815450643778
```

```python
df1=pd.read_csv('/content/marks.csv')
df1.head()
```

```python
pred=reg.predict(df1)
pred
```

```python
df1['marks']=pred
df1
```
