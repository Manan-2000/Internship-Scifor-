```python
import numpy as np
import matplotlib.pyplot as mpl
from keras.layers import Dense,Flatten
from keras.models import Sequential
from keras.utils import to_categorical
from keras import utils
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist

import warnings
warnings.filterwarnings('ignore')
```

```python
(x_train,y_train) , (x_test,y_test) = fashion_mnist.load_data()
```

```python
x_train.shape , y_train.shape , x_test.shape , y_test.shape
```

```python
fig, axes = mpl.subplots(ncols=10, sharex=False, sharey=True, figsize=(20, 4))

for i in range(10):
  axes[i].set_title(y_train[1])
  axes[i].imshow(x_train[i],cmap='gray')
  axes[i].get_xaxis().set_visible(True)
  axes[i].get_yaxis().set_visible(True)

mpl.show()
```

```python
class_names = ['0 T-shirt/top','1 Trouser','2 Pullover','3 Dress','4 Coat','5 Sandal','6 Shirt','7 Sneaker','8 Bag','9 Ankle boot']
print(class_names)
```

```python
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(7,activation='ReLU'))
model.add(Dense(10,activation='softmax'))
model.summary()
```

```python
utils.plot_model(model,show_shapes=True)
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'], optimizer='adam')
```

```python
mod = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```
